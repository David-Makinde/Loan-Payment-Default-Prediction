```python
data['form_field47'] = data['form_field47'].map({'charge':0, 'lending':1}).astype(np.float)

data['form_field14'] = data['form_field14'].astype(np.float)

data['form_field28'] = data['form_field28']*24.0
```

```python
# grouped by formfield 47 (i just tried it and luckily it really improved the prediction)
# although form field 1, 2, 6, 47 are really doing well on the features importance plot.see the plot at the end of the prediction

region = data.groupby('form_field47',).mean()
region.drop('default_status', 1, inplace = True)
cols = []
for i in region.columns:
    if i != 'form_field47':
        region[i+'_reg_mean_all'] = region[i]
        region.drop(i, 1, inplace = True)
        cols.append(i+'_reg_mean_all')

data = pd.merge(data, region, on='form_field47', how = 'left')
for col in cols: data[col+'_freq'] = data[col].map(data[col].value_counts().to_dict())/len(data)
```

# Data Modelling

```python
# get the dataset
def get_dataset():
    x,y = make_classification(n_samples=1000, n_features=20, n_informative=15, n_redundant=5, random_state=1)
    return x, y

# get a stacking ensemble of models
def get_stacking():
    level0 = list()
    #level0.append(('lr', LogisticRegression(max_iter = 100000, dual = False)))
    #level0.append(('knn', KNeighborsClassifier()))
    #level0.append(('cart', DecisionTreeClassifier()))
    level0.append(('svm', SVC()))
    #level0.append(('bayes', GaussianNB()))
    #level0.append(('rfc', RandomForestClassifier()))
    level0.append(('catboost', CatBoostClassifier(n_estimators=10000, max_depth=6, eval_metric='AUC', reg_lambda = 370)))
    level0.append(('xgboost', XGBClassifier(objective ='binary:logistic', 
                          eta = 0.99,
                          max_depth = 6, 
                          n_estimators = 5000,
                          reg_lambda = 500,
                          sub_sample = 0.8,
                          colsample_bytree = 0.8)))
    level0.append(('lightgbm', LGBMClassifier(objective =  'binary', 
                            metric= 'auc',
                            boosting_type= 'gbdt',
                            lambda_l1= 0.0004912993970392775,
                            lambda_l2= 9.424350138808432,
                            num_leaves= 24,
                            feature_fraction= 1.0,
                            bagging_fraction= 0.9540416539312312,
                            bagging_freq= 7,
                            min_child_samples= 100, n_estimators = 300)))
    
    level1 = CatBoostClassifier(n_estimators=10000, max_depth=6, eval_metric='AUC', reg_lambda = 370)
    
    model = StackingClassifier(estimators=level0, final_estimator=level1, cv=5)
    return model

# get a list of models to evaluate


def get_models():
    models = dict()
    #models['lr'] = LogisticRegression()
    #models['knn'] = KNeighborsClassifier()
    #models['cart'] = DecisionTreeClassifier()
    models['svm'] = SVC()
    #models['bayes'] = GaussianNB()
    #models['rfc'] = RandomForestClassifier()
    models['catboost'] = CatBoostClassifier(n_estimators=10000, max_depth=6, eval_metric='AUC', reg_lambda = 370)
    models['xgboost'] = XGBClassifier(objective ='binary:logistic', 
                          eta = 0.99,
                          max_depth = 6, 
                          n_estimators = 5000,
                          reg_lambda = 500,
                          sub_sample = 0.8,
                          colsample_bytree = 0.8)
    models['lightgbm'] = LGBMClassifier(objective =  'binary', 
                            metric= 'auc',
                            boosting_type= 'gbdt',
                            lambda_l1= 0.0004912993970392775,
                            lambda_l2= 9.424350138808432,
                            num_leaves= 24,
                            feature_fraction= 1.0,
                            bagging_fraction= 0.9540416539312312,
                            bagging_freq= 7,
                            min_child_samples= 100, n_estimators = 300)
    
    models['stacking'] = get_stacking()
    return models

# evaluate a given model using cross-validation
def evaluate_model(model, x, y):
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
    scores = cross_val_score(model, x, y, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')
    return scores
```

```python
# define dataset
x, y = get_dataset()

#get the models to evaluate
models = get_models()

# evaluate the models and store results
results, names = list(), list()
for name, model in models.items():
    scores = evaluate_model(model, x,y)
    results.append(scores)
    names.append(name)
    print('>%s %.sf (%.3f)' % (name, mean(scores), std(scores)))
    
# plot model performance for comparison
pyplot.boxplot(results, labels=names, showmeans=True)
pyplot.show()
```
